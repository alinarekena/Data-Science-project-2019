```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

specificity, specificity_err = (0.77,0.90,0.91),(0.00,0.00,0.00)
sensitivity, sensitivity_err = (0.82,0.66,0.74),(0.0194,0.0194,0.00)
accuracy, accuracy_err = (0.78,0.85,0.87),(0.00,0.00,0.00)

ind = np.arange(len(specificity))  # the x locations for the groups
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.barh(ind + width/2, specificity, width, xerr=specificity_err,
                label='Specificity', color='#90ee90', edgecolor='black')
rects2 = ax.barh(ind - width/2, sensitivity, width, xerr=sensitivity_err,
                label='Sensitivity', color='#add8e6',edgecolor='black')
rects3 = ax.barh(ind - width/0.67, accuracy, width, xerr=accuracy_err,
                label='Accuracy', color='#e6e6fa',edgecolor='black')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xticks(np.arange(0.0,1.21,0.2))
ax.set_xlabel('Performance (in decimal)')
ax.set_title('Model1: Imbalanced towards nB')
ax.set_yticks(ind)
ax.set_yticklabels(('GaussianNB','SVM,linear','Reference RF'))
ax.legend(loc='lower left')

for i, v in enumerate(specificity):
    ax.text(v + 0.01, i+0.2, str(v), color='black')

for i, v in enumerate(sensitivity):
    ax.text(v+0.07, i-0.15, str(v), color='black')
    
for i, v in enumerate(accuracy):
    ax.text(v+0.01, i-0.5, str(v), color='black')

fig.tight_layout()
plt.savefig('Model1-new.png')

plt.show()
```
